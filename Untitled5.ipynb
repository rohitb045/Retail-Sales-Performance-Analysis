```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("retail_sales.csv")
```

```python
# Display first 5 rows of data


print("Sample data:")
print(df.head())
```

```python
# Shape of dataset (rows, columns)

print("\nShape of dataset:", df.shape)
```

```python
# Display all column names
print("\nColumns in dataset:", df.columns)
```

```python
# Check for data types and missing values
print("\nData Info:")
print(df.info())
```

```python
# Quick statistics (mean, min, max, etc.)
print("\nStatistical summary:")
print(df.describe())
```

```python
# Check missing values
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Remove duplicate records if any
df.drop_duplicates(inplace=True)
```

```python
# Convert 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Verify data after cleaning
print("\nData after cleaning:")
print(df.head())
```

```python
# Extract month and year from date for trend analysis
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
# Verify new columns
print("\nAdded Month and Year columns:")
print(df.head())
```

```python
# A. Total Sales by Region
region_sales = df.groupby('Region')['Total_Sales'].sum().sort_values(ascending=False)
print("\nTotal Sales by Region:\n", region_sales)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=region_sales.index, y=region_sales.values, palette='viridis')
plt.title("Total Sales by Region")
plt.ylabel("Total Sales (₹)")
plt.xlabel("Region")
plt.show()
```

```python
#Top 5 Best-Selling Products

top_products = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False).head(5)
print("\nTop 5 Products by Sales:\n", top_products)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=top_products.index, y=top_products.values, palette='coolwarm')
plt.title("Top 5 Best-Selling Products")
plt.ylabel("Total Sales (₹)")
plt.xlabel("Product")
plt.show()
```

```python
# Monthly Sales Trend

monthly_sales = df.groupby('Month')['Total_Sales'].sum()
print("\nMonthly Sales Trend:\n", monthly_sales)
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o', color='orange')
plt.title("Monthly Sales Trend")
plt.ylabel("Total Sales (₹)")
plt.xlabel("Month")
plt.grid(True)
plt.show()
```

```python
#Category-wise Sales Distribution


category_sales = df.groupby('Category')['Total_Sales'].sum()
print("\nSales by Category:\n", category_sales)
```

```python
plt.figure(figsize=(5,5))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=90)
plt.title("Sales Distribution by Category")
plt.show()
```

```python
# Correlation Analysis

plt.figure(figsize=(6,4))
sns.heatmap(df[['Quantity','Unit_Price','Total_Sales']].corr(), annot=True, cmap='Blues')
plt.title("Correlation Heatmap")
plt.show()
```

```python
print(" INSIGHTS FROM ANALYSIS:")
print("1️ The region with the highest sales is:", region_sales.idxmax())
print("2️ Top-selling product is:", top_products.index[0])
print("3️ Peak sales month is:", monthly_sales.idxmax())
print("4️ Category contributing most revenue:", category_sales.idxmax())
print("5️ There is a strong correlation between Quantity and Total Sales.")
```
